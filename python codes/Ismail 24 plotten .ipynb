```python
# Read CSV file
df = pd.read_csv('/datc/parcel/notebooks/data/postnl/20200904_sample.csv', index_col=0, parse_dates=[1])
```

```python
# manipulate data
df['procesdag'] = df['procesdag'].where(df['procesdag'].dt.month == 12)
```

```python
#Plot
filter1 = df['cust_id'] == 'klant_1'
filter2 = df['procesdag'].dt.year == 2015
df_klant1 = df.where(filter1 & filter2).dropna()
fig, ax = plt.subplots()
ax.bar(df_klant1['procesdag'].dt.day, df_klant1['aantal_pakketten'])
ax.set(
    xlabel = 'Dag',
    ylabel = 'Aantal pakketten',
    title = 'Aantal pakketten per dag in December voor klant 1')
plt.show()
```
